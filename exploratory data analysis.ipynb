```python
nan_values=[]
for i in range(1,8):
    print(f"category{i}:")
    print("Unique values:"+str(len(df[f"category{i}"].unique())))
    print(("Nan values:"+str((df[f"category{i}"].isna()==True).sum())))
    nan_values.append((df[f"category{i}"].isna()==True).sum())
    print("-"*100)
```

```python
nan_values
nan_percentage=[str(int(i/df.shape[0]*100))+" %" for i in nan_values]
```

```python
import matplotlib.pyplot as plt
courses = list(range(1,len(nan_values)+1))
values = list(nan_percentage)
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
color = (0.2, # redness
         0.4, # greenness
         0.2, # blueness
         0.6 # transparency
         )
plt.bar(courses, values, color =color, 
        width = 0.4)
 
plt.xlabel("Category")
plt.ylabel("Percentage of nan values")
plt.title("Nan values")
plt.show()
```
